```python
#LIBRERIAS
import numpy as np
import pandas as pd
#libreria para usar elastic
from elasticsearch import Elasticsearch
#libreria para tratamiento de fechas
import datetime
```

```python
# Elastic search configuation
es = Elasticsearch("http://172.29.31.6:9200")
es.info()
```

```python
#creacion del indice
es.indices.create(index="peaky_pred_moli.del.sol")
```

```python
#leer el fichero
df = pd.read_csv("Prediccion_2021_46250048_PM10.csv", decimal=".",sep=";")
df.head()
```

#SUBIR A ELASTICSEARCH


```python
for i in df.index:
    #"lluvias","Temp","VelocVent","GEO"
    doc_1 = {"FECHA": df["FECHA"][i], "PM10": df["PM10"][i],"Prediccion": df["Prediccion"][i]}
    #####
    id = df["FECHA"][i]
    id = id.replace("-","")
    #id = "j" + id
    print(id)
    es.index(index="peaky_pred_avd.francia", id=id, body=doc_1)
print("OK")
```
