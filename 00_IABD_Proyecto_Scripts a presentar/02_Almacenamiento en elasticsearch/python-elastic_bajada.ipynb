```python
import numpy as np
import pandas as pd
#libreria para usar elastic
from elasticsearch import Elasticsearch
#libreria para tratamiento de fechas
import datetime
```

```python
# Elastic search configuation
es = Elasticsearch("http://172.29.31.6:9200")
es.info()
```

#bajada de datos de elastic en json

```python
#creando el operador para calcular los dias en este caso -1 dia
dia_delta = datetime.timedelta(days=1)
#calcular fecha actual
fecha_actual = datetime.date.today()
print(f"fecha actual: {fecha_actual}")
```

```python
df = pd.DataFrame()
final = False
fecha_bucle = fecha_actual
count = 0 #contador de datos guardados
n_error =0
while final == False:
    #le resto un dia al dia actual
    fecha_bucle = fecha_bucle + dia_delta
    try:
        #creamos un id personalizado para solicitar los datos en formato ao mas dia junto ej:20040522
        id = str(fecha_bucle)
        id = id.replace("-","")
        #peticion get a elastic
        res = es.get(index="prueba_moli2", id=id)
        #limpiamos la salida y lo pasamo a dataframe
        source = res["_source"]
        df1 = pd.json_normalize(source)
        df = pd.concat([df, df1], axis=0)
        print(df1["FECHA"])
        #contador de datos recopilados
        count = count +1
    except:
        n_error = n_error +1
    #if para detener bucle cuando llega a recaudar los ultimos x datos disponibes en elastic
    if count == 30:
        final = True
        print ("FIN")
```
