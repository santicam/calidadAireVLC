```python
import os
import pandas as pd
```

```python
# cargar dataset con todas las estaciones, definir la fecha como índice del dataframe

df = pd.read_csv('airWeather_2013-2019inter.csv', sep=';', index_col='FECHA', parse_dates=['FECHA'])

# Volver a recuperar la fecha como campo
df.insert(0, 'FECHA', df.index.strftime('%Y-%m-%d'))
```

```python
# Comprobar rango de fechas disponible para cada estación y registros disponibles
# df['FECHA'] = pd.to_datetime(df['FECHA'], format="%d/%m/%Y")
print(f'Nº de registros totales y rango de fechas por estación:')
print(f'{"NOM_RESTACION":=^30} {"Desde":=^10} {"Hasta":=^10} {"Nº reg":=^8}')
for esta in df['NOM_ESTACION'].unique():
  print(f'{esta:30} {str(df[df["NOM_ESTACION"]==esta]["FECHA"].min())[0:10]} {str(df[df["NOM_ESTACION"]==esta]["FECHA"].max())[0:10]} {df[df["NOM_ESTACION"]==esta]["NOM_ESTACION"].count():8}')
```
