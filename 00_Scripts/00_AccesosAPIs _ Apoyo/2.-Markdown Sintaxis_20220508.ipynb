# Índice de sintaxis Markdown #

En el lenguaje Markdown encontrarás tres tipos de elementos básicos que a su vez engloban el resto de la sintaxis. Considera esto una cheat sheet con la que guiarte.

1. Elementos de bloque
    - Párrafos y saltos de línea
    - Encabezados
    - Citas
    - Listas
    - Códigos de bloque
    - Reglas horizontales
2. Elementos de línea
    - Énfasis
    - Links o enlaces
    - Código
    - Imágenes
3. Elementos varios
    - Omitir Markdown
4. Multimarkdown 
    - Abreviaciones
    - Definiciones
    - Metadatos
    - Notas al pie de página
    - Referencias cruzadas
    - Tablas 

### **1. Elementos de bloque** ###
#### **Párrafos y saltos de línea** ####

Para generar un nuevo párrafo en Markdown simplemente separa el texto mediante una línea en blanco (pulsando dos veces intro)

Al igual que sucede con HTML, Markdown no soporta dobles líneas en blanco, así que si intentas generarlas estas se convertirán en una sola al procesarse.

Para realizar un **salto de línea dentro del mismo párrafo**, pulsar **dos veces la barra espaciadora antes de intro**.

Por ejemplo si quisieses escribir un poema Haiku quedaría tal que así:

«Andando con sus patitas mojadas,  
el gorrión  
por la terraza de madera»

Donde cada verso tiene dos espacios en blanco al final.

####  **Encabezados** ####

# Encabezado 1 # (1 x almohadilla )
## Encabezado 2 ## (2 x almohadilla )
### Encabezado 3 ### (3 x almohadilla )
#### Encabezado 4 #### (4 x almohadilla )
##### Encabezado 5 ##### (5 x almohadilla )
###### Encabezado 6 ###### (6 x almohadilla )

#### **Generar encabezados de dos niveles** ####

Esto sería un encabezado 1
===
Esto sería un encabezado 2
---

#### **Citas** ####

Cita : Símbolo mayor que (>) al comienzo del bloque de texto.

> Un país, una civilización se puede juzgar por la forma en que trata a sus animales.  — Mahatma Gandhi

Cita de varios párrafos: Símbolo mayor que (>) al comienzo de cada uno de ellos.

> Creo que los animales ven en el hombre un ser igual a ellos que ha perdido de forma extraordinariamente peligrosa el sano intelecto animal.
> Es decir, que ven en él al animal irracional, al animal que ríe, al animal que llora, al animal infeliz. — Friedrich Nietzsche

Concatenar citas (citas anidadas): Varios símbolos mayor que (>>).

> Esto sería una cita como la que acabas de ver.
> 
> > Dentro de ella puedes anidar otra cita.
> 
> La cita principal llegaría hasta aquí. 

Recuerda separar los saltos de línea con >, o >> si te encuentras dentro de la cita anidada; para crear párrafos dentro del mismo bloque de cita.

#### **Listas** ####

**Dos tipos.**

**Listas desordenadas:**

Se usan los símbolos (*) asteriscos, (-) guiones, o (+) suma. Son intercambiables

- Elemento de lista 1
- Elemento de lista 2
* Elemento de lista 3
* Elemento de lista 4
+ Elemento de lista 5
+ Elemento de lista 6

Listas anidadas : Añadir cuatro espacios en blanco antes del siguiente símbolo (*, -, +).

- Elemento de lista 1
- Elemento de lista 2
    - Elemento de lista 3
    - Elemento de lista 4
        - Elemento de lista 5
        - Elemento de lista 6

**Listas ordenadas**

Igual que las listas desordenadas pero usando númeración. Se pueden anidar o combinar.

1. Elemento de lista 1
2.  Elemento de lista 2
    - Elemento de lista 3
    - Elemento de lista 4
        1. Elemento de lista 5
        2. Elemento de lista 6

#### **Códigos de bloque** ####

Encerrar el párrafo entre dos líneas formadas por tres virgulillas (~).

~~~
Creando códigos de bloque.
Puedes añadir tantas líneas y párrafos como quieras.  
~~~


#### **Reglas horizontales** ####

Separacion visual de secciones.

Se utilizan dentro de una linea en blanco incluyendo tres de los siguientes elementos: asteriscos, guiones, o guiones bajos.

Es decir

***
---
___

También puedes separarlos mediante un espacio en blanco por pura estética.

* * *
- - -
_ _ _


### **2. Elementos de línea** ###
#### **Énfasis (negritas y cursivas)** ####
Markdown utiliza asteriscos o guiones bajos para enfatizar.
Simplemente tendrás que envolver palabras o textos en éstos símbolos para conseguir cursivas o negritas.
| Markdown | Sintaxis |
| -- | -- |
| *cursiva* | Abre y cierra con un (1) asterisco (*) |
| _cursiva_ | Abre y cierra con un (1) guión bajo (_) |
| **negrita** | Abre y cierra con dos (2) asteriscos (**) |
| __negrita__ | Abre y cierra con dos (2) guiones bajos (__) |  
---
---
| Markdown | Sintaxis |
| -- | -- |
| ***negritacursiva*** | Abre y cierra con tres (3) asteriscos (***) |
| ___negritacursiva___ | Abre y cierra con tres (3) guiones bajos (___) |

#### **Links o enlaces** ####

**Links o enlaces en línea**

Texto del enlace entre **[] corchetes** y link entre **() paréntesis**.
| Markdown | Sintaxis |
| -- | -- |
| [enlace en línea] | (http://www.limni.net)|


**Links o enlaces como referencia**
Los enlaces se definen en otra parte del documento para permitir maś legibilidad.

SINTAXIS :

[nombre que quieres darle a tu enlace][nombre de tu referencia]
[nombre de tu referencia]: http:www.tuenlace.com

Ejemplo.

Me llamo Jeps Tunar y tengo un blog sobre [productividad mac][blog].

En dicha [web][blog] recopilo artículos sobre todo lo relacionado con automatización, gestión y eficiencia.

[blog]: http://limni.net/blog/

La referencia [blog] puede estar incluida en cualquier parte del documento y estas las recopilaremos agrupadas en un mismo lugar. Las referencias no aparecen en el texto mostrado.

**Links automáticos**

Son necesarios cuando se quiere mostrar una URL completa, y no un enlace enmascarado.

SINTAXIS : 
Se generan rodeandolos con los símbolos menorque (<) y mayor que (>)

<http://www.limni.net>

#### **Código** ####
Secciones donde mostrar código de otro lenguaje, atajos de teclado...

Dos alternativas.

**Código puro**

Colocaremos el texto entre acentos abiertos (``) o con HTML mediante la etiqueta code

<code>Aqui el código con la etiqueta code</code>

`Aquí entre acentos abiertos`


**Texto preformateado**

Comenzar el párrafo con cuatro espacios en blanco

    Esto es una línea de código con cuatros espacios en blanco al principio ( indentación de 4 espacios)

#### **Imágenes** ####

SINTAXIS:

![Texto alternativo(ALT)](/ruta/a/la/imagen.jpg)

El texto alternativo es lo que se mostraría si la carga de la imagen fallase.

![Texto alternativo](/ruta/a/la/imagen.jpg "Título alternativo")

El título alternativo se muestra durante un rollover de ratón.



SINTAXIS 2:

![nombre de la imagen][img1]

![nombre de la imagen2][img2] 

Estos serían los links de referencia al igual que se ha visto en la sintaxis de links.

[img1]: /ruta/a/la/imagen.jpg "Título alternativo"

[img2]: /ruta/a/la/imagen2.jpg "Título alternativo"

### **3. Elementos varios** ###

#### **Omitir Markdown** ####

Cómo escribir ciertos símbolos como * asteriscos o _ guiones bajos, sin que Markdown los interprete en su sintaxis , sino como símbolos a mostrar.

SINTAXIS : Se trata de "escapar" los símbolos mediante la barra invertida (\\) delante del símbolo.

- \\  barra invertida
- \`  acento invertido
- \*  asterisco
- \_  guión bajo
- \{} llaves
- \[] corchetes
- \() paréntesis
- \#  almohadilla
- \+  símbolo de suma
- \-  guión
- \.  punto
- \!  exclamación

### **4. Multimarkdown** ###

#### **Abreviaturas** ####
Sintaxis:

MMD ofrece muchas más posibilidades que MD.

*[MMD]: Abreviatura para Multimarkdown

*[MD]: Abreviatura para Markdown

Es igual que crear links de referencia, solo que precedidos por un asterisco.

Se visualiza con un rollover de ratón.Para descubrir cuál es el resultado final, no tienes más que dejar el puntero del ratón sobre la palabra o .
ofrece muchas más posibilidades que .

#### **Definiciones** ####
Sintaxis: Dos puntos y un espacio antes de escribir

Markdown

: Herramienta de conversión de texto plano a HTML.

: Lenguaje de marcado que facilita la escritura de documentos y artículos web.

Multimarkdown

: Lenguaje de marcado que añade características a las carencias de su hermano Markdown (tablas, notas al pie de página, listas de definiciones como esta…)

Esto se transforma en:

Markdown

Multimarkdown

#### **Metadatos** ####
Sintaxis : Añadirlos en la parte superior del documento:

Title: Guía para la utilización de Markdown

Author: Pep Jepse

Date: 12 de noviembre de 2015

A tener en cuenta :
- No todas las aplicaciones soportan este añadido. En las que lo admiten desaparecerá al transformarse a un nuevo documento (por ejemplo, PDF)
- No debe haber ningún espacio en blanco por encima de los metadatos.
- Un salto de línea en blanco (nuevo párrafo) da por terminado el bloque de metadatos. Para poner varios autores, hay que hacer un salto de línea simple (espacio espacio intro).
- Funcionan en inglés.

Algunos de los Metadatos soportados por Multimarkdown ( en total hay más de 30 ):
- Address: Incluirá la dirección del autor (Truco : espacio espacio intro para añadir varias líneas)
- Author: Incluirá el nombre del autor.
- Affiliation: Incluirá las organizaciones con las que el documento puede estar afiliado (compañías, universidades…)
- Comment: Incluirá un campo para añadir comentarios acerca de la obra.
- Copyright: Incluirá un copyright o tipo de licencia en tu documento.
- Date: Incluirá fecha de creación del documento.
- Email: Incluirá el email del autor.
- Keywords: Incluirá las palabras clave que definen el documento, separadas por comas.
- Subtitle: Incluirá el subtítulo oficial del documento.
- Title: Incluirá el título oficial del documento.
- Web: Incluirá la URL del autor.

#### **Notas al pie de página** ####

Sintaxis:
[^nombreDeNota]

Ejemplo:
Esto es un texto normal que contiene una nota [^nombreDeNota]. Se escribir tanto como se necesite.
Incluso crear varios párrafos, ya que la nota siempre irá al final del documento. En este caso, al hacer click se dirige al final del artículo.

[^nombreDeNota]: Aquí irá el texto de tu nota.

#### **Referencias cruzadas** ####
Sirven para añadir links a partes específicas del documento.

Sintaxis :
Entre llaves el nombre de dicho enlace {#EnlaceReferencia}.

Encabezado que se referencia {#EncabezadoReferencia}

Para enlazarlo desde cualquier texto se hace mediante el #NombreDeLaReferencia.

Desde aquí con [este enlace](#referencia) se desplazará al encabezado de referencias cruzadas.


#### **Tablas** ####

Sintaxis :

Cabecera de cada columna : Se colocarán entre barras verticales (|):

| Primera columna | Segunda columna | Tercera columna |

Línea adicional : Debajo del encabezado para cerrarlo. Una línea como la del encabezado que simplemente incluirá dos guiones (--).

| -- | -- | -- |

Contenido de la tabla : Se crean exactamente igual que la línea de cabecera incluyendo en cada línea el contenido a mostrar.


| Primera columna | Segunda columna | Tercera columna |
| -- | -- | -- |
| Contenido 1-1 | Contenido 1-2 | Contenido 1-3 |
| Contenido 2-1 | Contenido 2-2 | Contenido 2-3 |
| Contenido 3-1 | Contenido 3-2 | Contenido 3-3 |
