```python
# Librerías
import pandas as pd
import numpy as np
```

```python
# path = '/content/drive/MyDrive/datoscsv/'
# path local para probar:
path = 'datoscsv/'
#df = pd.read_csv(path + 'airWeather_2013-2019_v2.csv', sep=';', index_col='FECHA', parse_dates=['FECHA'])
df = pd.read_csv(path + 'Valencia_TODASEstaciones_CON_Meteo_1994_2022.csv', sep=';', index_col='FECHA', parse_dates=['FECHA'])

# Recuperar la fecha como campo
df.insert(0, 'FECHA', df.index.strftime('%Y-%m-%d'))
df
```

```python
# % de datos faltantes
print ('Antes de interpolar')
df.isnull().sum()/len(df)*100
```

```python
# Interpolar valores faltantes en base a índice de fechas, siempre que el hueco no sea de más de 4 valores NaN

df = df.interpolate(method='time', limit=3, limit_direction='both', limit_area='inside')
```

```python
# % de datos faltantes
print ('Después de interpolar')
df.isnull().sum()/len(df)*100
```

```python
# Escribir csv depués de interpolar
df.to_csv(path + 'Valencia_TODASEstaciones_CON_Meteo_1994_2022_v3.csv', header=True, sep=';', index=False)
```
