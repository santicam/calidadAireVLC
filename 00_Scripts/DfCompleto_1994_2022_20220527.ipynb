```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
#Nombre del directorio de datos
PATH = "datoscsv/"
pathname = os.path.join(PATH)
print(pathname)
```

```python
maindf = pd.read_csv(pathname + "SoloVLC_Principales.csv",sep=";",decimal=".")
```

```python
maindf.drop([ 'Pb','Ni','IcdP','UVB','NH3','R.Sol.','Pres.','C8H10','H.Rel.'] , axis=1, inplace= True)
maindf.drop([ 'HMN', 'HNM','C6H6','C7H8','As', 'Cd','H2S','HCT','HMN', 'FA', 'DahA', 'CH4', 'BaA'] , axis=1, inplace= True)
maindf.drop([ 'BkFA', 'BjFA', 'BbFA', 'BaP','PST'] , axis=1, inplace= True)
```

```python
maindf.isnull().sum()/len(maindf)*100
```

```python
maindf.columns
```

```python
maindf['COD_ESTACION'] = maindf['COD_ESTACION'].astype(str)
maindf['NOM_ESTACION'] = maindf['NOM_ESTACION'].astype(str)
# Convertir la fecha a formato datetime
#maindf['FECHA'] = pd.to_datetime(maindf['FECHA'], format="%d-%m-%Y")
maindf['FECHA'] = pd.to_datetime(maindf['FECHA'], format="%Y-%m-%d")
```

```python
maindf.head()
```

```python
maindf.tail()
```

```python
columnas = maindf.columns.to_list()
borra1 = "COD_ESTACION"
borra2 = 'NOM_ESTACION'
borra3 = 'FECHA'
columnas = list(filter((borra1).__ne__,columnas))
columnas = list(filter((borra2).__ne__,columnas))
columnas = list(filter((borra3).__ne__,columnas))
```

```python
for col in columnas:
    maindf[col] = maindf[col].replace(',','.', regex=True)
    maindf[col] = maindf[col].apply(pd.to_numeric)
```
