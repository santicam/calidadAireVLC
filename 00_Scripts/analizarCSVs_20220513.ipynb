```python
# Librerias necesarias
 
import os
import pandas as pd
```

```python
# Obtener lista de archivos CSV

import os
listaCsv = []
for fichero in os.listdir():
    if fichero.endswith(".csv"):
        listaCsv.append(fichero)
print(listaCsv)
```

```python
# Juntar todos los csv en un dataframe único para analizar datos

# Crear dataframe vacio
df = pd.DataFrame()
# Crear una lista de columnas vacía
columnas = []
```

```python
# Cargar todos los archivos csv de las distintas estaciones
for fichero in listaCsv:
    dfaux = pd.read_csv(fichero, sep=';', decimal='.')
    # Comprobar si hay nuevas columnas
    print(f'>>>>>>>>> {fichero} ', end =',') 
    for c in dfaux.columns:
        if c not in columnas:
            print(f' {c}', end= ',')
            columnas.append(c)
    print()
    df = pd.concat([df, dfaux])
```

```python
for c in df.columns:
    porcentaje = df[c].isnull().sum()/len(df)*100
    if porcentaje >= 80.0 and c != 'Precip.':
        print(f'Se va a eliminar: {c} {porcentaje}')
        df = df.drop([c], axis=1 )
```
